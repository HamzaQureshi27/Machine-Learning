```python
# 3. Write a Python program to count the occurrences of each word in a given sentence.

sentence = input("Enter a sentence: ")

# Convert to lowercase and split into words
words = sentence.lower().split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("Word occurrences:")
for word, count in word_count.items():
    print(word, ":", count)
```

```python
# 5. Write a Python program to add 'ing' at the end of a given string (length should be at least 3).If the given string already ends with 'ing' then add 'ly' instead If the string length of the given string is less than 3, leave it unchanged

string = input("Enter a string: ")

if len(string) < 3:
    result = string
elif string.endswith("ing"):
    result = string + "ly"
else:
    result = string + "ing"

print("Result:", result)
```

```python
# 6. Write a Python program to find the first appearance of the substring 'not' and 'poor' from a given string, if 'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'. Return the resulting string

string = input("Enter a string: ")

not_pos = string.find("not")
poor_pos = string.find("poor")

if not_pos != -1 and poor_pos != -1 and poor_pos > not_pos:
    result = string[:not_pos] + "good" + string[poor_pos + 4:]
else:
    result = string

print("Result:", result)
```

```python
# 7. Program to find Greatest Common Divisor of two numbers.
# For example, the GCD of 20 and 28 is 4 and GCD of 98 and 56 is 14.

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

gcd = 1

for i in range(1, min(num1, num2) + 1):
    if num1 % i == 0 and num2 % i == 0:
        gcd = i

print("GCD is:", gcd)
```

```python
# 8. Write a Python program to check whether a list contains a sublist.

def contains_sublist(lst, sublst):
    n, m = len(lst), len(sublst)

    for i in range(n - m + 1):
        if lst[i:i + m] == sublst:
            return True
    return False

# Example usage
main_list = [1, 2, 3, 4, 5]
sub_list = [3, 4]

if contains_sublist(main_list, sub_list):
    print("Sublist found!")
else:
    print("Sublist NOT found.")
```

```python
# 13. Write a Python program to sort a dictionary (ascending /descending) by value.

my_dict = {'apple': 40, 'banana': 10, 'mango': 30, 'orange': 20}

# Ascending order by value
ascending = dict(sorted(my_dict.items(), key=lambda item: item[1]))

# Descending order by value
descending = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))

print("Ascending:", ascending)
print("Descending:", descending)
```

```python
"""
15. Given a number n, write a python program to make and print the list of Fibonacci series up to n.
Input : n=7
Hint : first 7 numbers in the series
Expected output :
First few Fibonacci numbers are 0, 1, 1, 2, 3, 5, 8, 13
"""

n = int(input("Enter a number: "))

# Initialize the first two Fibonacci numbers
fib_series = [0, 1]

# Generate Fibonacci series up to n terms
for i in range(2, n + 1):
    next_num = fib_series[i - 1] + fib_series[i - 2]
    fib_series.append(next_num)

print("First few Fibonacci numbers are:", ", ".join(map(str, fib_series)))
```

```python
"""
16. Counting the frequencies in a list using a dictionary in Python.
Input : [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,4, 4, 4, 2, 2, 2, 2]
Expected output : 1 : 5 , 2 : 4 , 3 : 3 , 4 : 3 , 5 : 2
"""

# Input list
numbers = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1, 4, 4, 4, 2, 2, 2, 2]

# Create an empty dictionary to store frequencies
freq_dict = {}

# Count frequencies
for num in numbers:
    if num in freq_dict:
        freq_dict[num] += 1
    else:
        freq_dict[num] = 1

# Print frequencies
for key, value in freq_dict.items():
    print(f"{key} : {value}", end=" , ")
```

```python
"""
17. Write a python program using function to find  the sum of odd series and even series
Odd series:  12/ 1!  +   32/ 3!   +  52/ 5!+……n
Even series: 22/ 2!  +   42/ 4!   +  62/ 6!+……n
"""

import math

# Function to calculate sum of odd series
def sum_odd_series(n):
    total = 0
    for i in range(1, n+1, 2):  # odd numbers
        total += (i**2) / math.factorial(i)
    return total

# Function to calculate sum of even series
def sum_even_series(n):
    total = 0
    for i in range(2, n+1, 2):  # even numbers
        total += (i**2) / math.factorial(i)
    return total

# Input n
n = int(input("Enter the maximum term n: "))

odd_sum = sum_odd_series(n)
even_sum = sum_even_series(n)

print("Sum of odd series:", odd_sum)
print("Sum of even series:", even_sum)
```

```python
# 18. Python Program to Find Factorial of Number Using Recursion

# Recursive function to calculate factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Input from user
num = int(input("Enter a number: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print(f"Factorial of {num} is: {factorial(num)}")
```

```python
# 19. Write a Python function that takes a list and returns a new list with unique elements of the first list.

# Function to return unique elements
def unique_elements(lst):
    unique_list = []
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

# Example usage
original_list = [1, 2, 2, 3, 4, 4, 5, 1]
new_list = unique_elements(original_list)

print("Original list:", original_list)
print("List with unique elements:", new_list)
```

```python
"""
20. Mini project :
Problem Statement : Password Generator
Make a program to generate a strong password using the input given by the user. To generate a password,
randomly take some words from the user input and then include numbers, special characters and capital
letters to generate the password. Also, keep a check that password length is more than 8 characters.
Note: Include Exception handling wherever required. Also, make a ‘User’ class and store the details like user
id, name and password of each user as a tuple.
"""

import random
import string

# User class to store user details
class User:
    def __init__(self, user_id, name, password):
        self.user_id = user_id
        self.name = name
        self.password = password

    def get_user_details(self):
        return (self.user_id, self.name, self.password)

# Function to generate strong password
def generate_password(user_input):
    try:
        if not user_input.strip():
            raise ValueError("Input cannot be empty.")

        words = user_input.split()
        if len(words) == 0:
            raise ValueError("No valid words found in input.")

        # Randomly pick 2-3 words
        num_words = min(len(words), random.randint(2, 3))
        selected_words = random.sample(words, num_words)
        password = ''.join(selected_words)

        # Add capital letters
        password = ''.join([c.upper() if random.choice([True, False]) else c for c in password])

        # Add numbers and special characters
        num_digits = random.randint(1, 3)
        num_special = random.randint(1, 2)
        password += ''.join(random.choices(string.digits, k=num_digits))
        password += ''.join(random.choices(string.punctuation, k=num_special))

        # Ensure password length > 8
        while len(password) < 8:
            password += random.choice(string.ascii_letters + string.digits + string.punctuation)

        # Shuffle the password
        password = ''.join(random.sample(password, len(password)))

        return password
    except Exception as e:
        print("Error in generating password:", e)
        return None

# Main program
try:
    user_id = input("Enter user ID: ").strip()
    name = input("Enter your name: ").strip()
    user_input = input("Enter some words to help generate password: ").strip()

    password = generate_password(user_input)

    if password:
        # Create user object
        user = User(user_id, name, password)
        print("\nUser created successfully!")
        print("User details (id, name, password):", user.get_user_details())

except Exception as e:
    print("Error:", e)

```
